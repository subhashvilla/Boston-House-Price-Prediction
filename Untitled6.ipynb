```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
boston_data = {
    'data': data,
    'target': target,
    'feature_names': ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'],  # Add feature names
    'DESCR': "Boston House Prices dataset\n\nThis dataset contains information about housing in Boston, including features like crime rate, average number of rooms, and median value of owner-occupied homes."  # Add a brief description
}

# Print the data
print(boston_data)
```

```python
df_x = pd.DataFrame(data, columns=boston_data['feature_names'])
df_y = pd.DataFrame(target, columns=['price'])
```

```python
# TO GET SOME STATISTICS FROM DATA SET LIKE
```

```python
df_x.describe()
```

```python
# initilize linear regression model
reg = linear_model.LinearRegression()
```

```python
# split and data 67% training and 33%testing
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.2,random_state=80)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# training model with our training data
reg.fit(x_train,y_train)
```

```python
# print the coefficientst for each feature
print(reg.coef_) #f(x) = mx+b = y
```

```python
y_pred = reg.predict(x_test)
print(y_pred)
```
